```python
from matplotlib import pyplot as plt
import numpy as np
import os
from os import path
```

```python
classNames = [f.name for f in os.scandir("lettersDataset_final\\test") if f.is_dir()]
classCountsTrain = [len([name for name in os.listdir(f.path)]) for f in os.scandir("lettersDataset_final\\train") if f.is_dir()]
classCountsTest = [len([name for name in os.listdir(f.path)]) for f in os.scandir("lettersDataset_final\\test") if f.is_dir()]
classCounts = [x + y for (x, y) in zip(classCountsTest, classCountsTrain)]
```

```python
x = np.arange(len(classNames))  # the label locations

counts = [classCounts, classCountsTrain, classCountsTest]

fig, ax = plt.subplots(nrows=1,ncols=len(counts))

fig.set_figwidth(20)

for (i, data) in enumerate(counts):
    ax[i].bar(x, data)

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax[i].set_xticks(x)
    ax[i].set_xticklabels(classNames)


ax[0].title.set_text("Total counts per class")
ax[1].title.set_text("Train counts per class")
ax[2].title.set_text("Test counts per class")
```
